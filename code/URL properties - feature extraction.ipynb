```python
benign_url_data = pd.read_csv("datasets/alexa_top1m/top-1m.csv", header=None, names=["SrNo","domain"])
benign_url_data.head()
```

```python
benign_url_data['URL'] = 'http://' + benign_url_data['domain'].astype(str)
benign_url_data.head()
```

```python
# https://docs.python.org/3/library/urllib.parse.html

from urllib.parse import urlparse
def getURLattributes(url_str) : 
    parsed_uri = urlparse(url_str)
    #result = '{uri.scheme}://{uri.netloc}/'.format(uri=parsed_uri)
    #print(result)
    #return [parsed_uri.scheme, parsed_uri.netloc, parsed_uri.path,  parsed_uri.params, parsed_uri.query, parsed_uri.fragment]
    return  pd.Series({'scheme':parsed_uri.scheme, 'netloc':parsed_uri.netloc, 'params':parsed_uri.params, 'query': parsed_uri.query, 'fragment':parsed_uri.fragment})
```

```python
#print(getURLattributes("https://docs.python.org/3/library/urllib.parse.html"))

#benign_url_data['URL'].map(getURLattributes).head()

benign_url_data = benign_url_data.merge(benign_url_data['URL'].apply(getURLattributes), left_index=True, right_index=True)
```

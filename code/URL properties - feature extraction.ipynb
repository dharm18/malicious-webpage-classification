```python
import pandas as pd
```

```python
benign_url_data = pd.read_csv("datasets/alexa_top1m/top-1m.csv", header=None, names=["SrNo","domain"])
benign_url_data.head()
```

```python
benign_url_data['URL'] = 'http://' + benign_url_data['domain'].astype(str)
benign_url_data.head()
```

```python
benign_url_data = (benign_url_data[["URL"]])
benign_url_data.head()
```

```python
# read malicious data
mal_url_data = pd.read_csv("datasets/phishtankurl.csv")
print(mal_url_data.head())
```

```python
mal_url_data = (mal_url_data[["url"]])
mal_url_data.head()
```

```python
mal_url_data.columns = ["URL"]
mal_url_data.head()
```

```python
mal_url_data["class"] = 'Malicious'
benign_url_data["class"] = 'Benign'
mal_url_data.head()
benign_url_data.head()
url_data = benign_url_data.append(mal_url_data, ignore_index=True)
url_data.info()
```

```python
url_data["class"] = url_data["class"].astype('category')
url_data["URL"] = url_data["URL"].astype(str)
url_data.info()
```

```python
url_data['URL_length'] = url_data['URL'].str.len()
url_data.head()
```

```python
# https://docs.python.org/3/library/urllib.parse.html

from urllib.parse import urlparse
def getURLattributes(url_str) : 
    parsed_uri = urlparse(url_str)
    #result = '{uri.scheme}://{uri.netloc}/'.format(uri=parsed_uri)
    #print(result)
    #return [parsed_uri.scheme, parsed_uri.netloc, parsed_uri.path,  parsed_uri.params, parsed_uri.query, parsed_uri.fragment]
    return  pd.Series({'scheme':parsed_uri.scheme, 'domain':parsed_uri.netloc, 'path':parsed_uri.path, 'params':parsed_uri.params, 'query': parsed_uri.query, 'fragment':parsed_uri.fragment})
```

```python
#print(getURLattributes("https://docs.python.org/3/library/urllib.parse.html"))

#benign_url_data['URL'].map(getURLattributes).head()

url_data = url_data.merge(url_data['URL'].apply(getURLattributes), left_index=True, right_index=True)
```

```python
url_data.to_csv('url_data_with_attr.csv', encoding='utf-8')
```

```python
url_data['domain_length'] = url_data['domain'].str.len()
url_data.head()
```

```python
url_data.tail()
```

```python
url_data['domain_token'] = url_data['domain'].apply(lambda s: (s.split('.')))
url_data.tail()
```

```python
url_data['domain_token_count'] = url_data['domain_token'].apply(len)
url_data.tail()
```

```python
url_data['domain_tokens_length'] = url_data['domain_token'].apply(lambda s: [len(i) for i in s])
url_data.tail()
```

```python
from statistics import mean
url_data['avg_domain_tokens_length'] = url_data['domain_tokens_length'].apply(lambda s: mean(s))
url_data.tail()
```

```python
url_data['longest_domain_tokens_length'] = url_data['domain_tokens_length'].apply(lambda s: max(s))
url_data.tail()
```

```python
url_data['dash_in_domain_count'] = url_data['domain'].apply(lambda s: str.count(s, '-'))
url_data.tail()
```

```python
url_data['path_length'] = url_data['path'].str.len()
url_data.tail()
```

```python
url_data.to_csv('url_data_with_attr2.csv', encoding='utf-8')
```

```python
url_data['numeric_count'] = url_data['domain'].apply(lambda s: sum(c.isdigit() for c in s)) #sum of numerics in a domain
url_data.tail()
url_data['numeric_ratio'] = url_data['numeric_count']/url_data['domain_length']
url_data.tail()
```

```python
url_data['path_numeric_count'] = url_data['path'].apply(lambda s: sum(c.isdigit() for c in s)) #sum of numerics in a domain
url_data.tail()
url_data['path_numeric_ratio'] = url_data['path_numeric_count']/url_data['path_length']
url_data.tail()
```

```python
import re

url_data['path_tokens'] = url_data['path'].apply(lambda s: (re.compile(r'[\:/?=\-&]+',re.UNICODE).split(s)))
url_data['path_tokens_length'] = url_data['path_tokens'].apply(lambda s: [len(i) for i in s])
url_data['avg_path_tokens_length'] = url_data['path_tokens_length'].apply(lambda s: mean(s))
url_data['longest_path_tokens_length'] = url_data['path_tokens_length'].apply(lambda s: max(s))
url_data.tail()
```

```python
#url_data['symbols_in_path_count'] = url_data['path'].apply(lambda s: sum(re.compile('[@_!#$%^&*()<>?/\|}{~:]').search(c) for c in s)) #sum of special chars in a domain
url_data['symbols_in_path_count'] = url_data['path'].apply(lambda s: len(re.sub('[\w]+' ,'', s)))
url_data.tail()
```

```python
url_data.to_csv('url_data_with_attr3.csv', encoding='utf-8')
```

```python
url_data = pd.read_csv('url_data_with_attr3.csv', index_col=0, encoding='utf-8')
```

```python
#EDA
print(url_data.shape)
print(url_data["class"].value_counts())
```

```python
url_data_model = url_data[["class","URL_length","domain_length","path_length","numeric_ratio","path_numeric_ratio","avg_path_tokens_length","longest_path_tokens_length","symbols_in_path_count"]]
url_data_model.describe()
```

```python
# Create features array and the response
y = url_data_model['class'].values
X = url_data_model.drop('class', axis=1).values
```

```python
url_data_model.head()
```

```python
# Import necessary modules
from sklearn.preprocessing import Imputer
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

# Setup the pipeline steps: steps
steps = [('imputation', Imputer(missing_values='NaN', strategy='most_frequent', axis=0)),
        ('SVM', SVC())]

# Create the pipeline: pipeline
pipeline = Pipeline(steps)

# Create training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Fit the pipeline to the train set
pipeline.fit(X_train,y_train)

# Predict the labels of the test set
y_pred = pipeline.predict(X_test)


```

```python
from sklearn.metrics import classification_report
# Compute metrics
print(classification_report(y_test, y_pred))
```
